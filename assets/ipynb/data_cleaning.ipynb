```python
# Importing the dependencies
import pandas as pd
```

```python
# Air pollution Data in the United States
air_pollution = pd.read_csv('../data/2020_Airpollution.csv')
air_pollution
```

```python
# Crime and Incarceration in the US
crime_data = pd.read_csv('../data/crime_and_incarceration_by_state.csv')
crime_data
```

```python
# Dropping the federal row
crime_data1 = crime_data.drop(0)
crime_data1
```

```python
crime_data2014 = crime_data1[crime_data1['year'] == 2014]
crime_data2014
```

```python
crime_data2014 = crime_data2014.drop(663)
crime_data2014
```

```python
crime_data2014 = crime_data2014.reset_index(drop=True)
crime_data2014
```

```python
crime_data2014 = crime_data2014.drop(columns=['index'])
crime_data2014
```

```python
# Obtaining a list of columns
crime_data2014.columns
```

```python
# Filtering for relevant columns
crime_data2014 = crime_data2014.drop(columns=['includes_jails', 'year','crime_reporting_change', 'crimes_estimated','violent_crime_total', 'murder_manslaughter', 'rape_legacy',
       'rape_revised', 'robbery', 'agg_assault', 'property_crime_total',
       'burglary', 'larceny', 'vehicle_theft'])
crime_data2014
```

```python
# renaming the columns
crime_data2014 = crime_data2014.rename(columns={'jurisdiction':'state'})
crime_data2014
```

```python
# Reading the health data frame
healh_data = pd.read_csv('../data/data.csv')
healh_data
```

```python
healh_data['state'] = healh_data['state'].str.lower()
```

```python
healh_data = healh_data.reset_index(drop=True).drop(columns=['id'])
healh_data
```

```python
crime_data2014['state'] = crime_data2014['state'].str.lower()
crime_data2014
```

```python
air_pollution['state'] = air_pollution['state'].str.lower()
air_pollution
```

```python
# list of Data frames to merge are 
# healh_data
# air_pollution
# crime_data2014
```

```python
merged_df = pd.merge(healh_data, air_pollution, how='inner', on='state')
merged_df
```

```python
merged_df1 = pd.merge(merged_df,crime_data2014, how='inner', on='state')
merged_df1
```

```python
# Exporting the dataset to csv for Data Visualisation
merged_df1.to_csv('../data/data1.csv')
```
